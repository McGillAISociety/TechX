# Data Visualization Demo

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Line plot**

```python
x = [x*0.1 for x in range(100)]
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

**Scattered plot**

```python
plt.scatter(x, y)
plt.show()
```

**Bar Chart**

```python
u = ['Red', 'Blue', 'Green']
v = [1, 2, 3]
```

```python
plt.bar(u, v)
plt.show()
```

Change colors

```python
plt.bar(u, v, color=('r', 'b', 'g'))
plt.show()
```

**How about adding having two graphs in the figures?**

```python
plt.figure(1)
plt.plot(x, y)
plt.figure(2)
plt.bar(u, v)
plt.show()
```

**You can even have them appear on the same graph**

```python
x = [x*0.1 for x in range(100)]
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y, 'b')
plt.plot(x, y2, 'r--')
plt.show()
```

**Add titles and legens!**

```python
plt.title('Demo Graph')
plt.xlabel('x-axis')
plt.ylabel('cos/sin')
plt.plot(x, y, 'b')
plt.plot(x, y2, 'r--')
plt.show()
```
